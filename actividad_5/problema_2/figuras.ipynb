## Figuras para la presentaci√≥n

### 1) Transferencia

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


R      = np.loadtxt('datos_2805/R_p2_fc8_s12_o3_sN.txt')
T      = np.loadtxt('datos_2805/T_p2_fc8_s12_o3_SN.txt')

frecs      = np.logspace(np.log10(40), np.log10(50000), 100, dtype=int)
frecs_fino = np.arange(82, 131, 1, dtype=int)

Transferencia = 2*R/0.1


fig, ax  = plt.subplots(1, figsize =(18,5),facecolor='white')

ax.plot(frecs,  np.log10(Transferencia), 'ok')
ax.set_xscale('log')
ax.set_xlabel('Frecuencia [Hz]', fontsize=15)
ax.set_ylabel('$V_{out}/V_{in}$ [dB]', fontsize=15)
ax.set_title('Transferencia', fontsize=20)
ax.grid()
ax.tick_params(labelsize=15)
plt.show()
```

### 2) Fase

```python


fig, ax  = plt.subplots(1, figsize =(18,5),facecolor='white')

ax.plot(frecs, T*2*np.pi / 360, 'ok')
ax.set_xscale('log')
ax.set_xlabel('Frecuencias [Hz]', fontsize=15)
ax.set_ylabel('Fase $\phi$ [Rad]', fontsize=15)
ax.set_title('Fase', fontsize=20)
ax.set_ylim([-183*2*np.pi / 360, -170*2*np.pi / 360])
ax.plot([10, 1e5], [-np.pi, -np.pi], linestyle="dashed", color="red", label="$-\pi$")
ax.set_xlim([10**(1.4),1e5])
ax.grid()
ax.legend(fontsize=15)
ax.tick_params(labelsize=15)
plt.show()
```
