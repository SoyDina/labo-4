$$v(t) = |cos(2\pi ft)|$$
$$v(t) = |cos(2\pi ft)|+0.2  \quad f = 45Hz$$
$$v(t) = |cos(2\pi ft)|^2 +0.2 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
import matplotlib
%matplotlib inline
#matplotlib.use('Qt5Agg')
#Qt5Agg
N =500
tmax = 0.2
f = 45
t = np.linspace(0,tmax,N)
v1 = np.abs(np.cos(2*np.pi*f*t))
v2 = np.abs(np.cos(2*np.pi*f*t)) +0.2
v3 = (np.abs(np.cos(2*np.pi*f*t)))**2 +0.2

tstep = tmax/N
fsamp = 1/tstep

y1 = np.fft.fft(v1)
frec = np.linspace(0,fsamp, N)


plt.figure(figsize = (10,4))
plt.plot(t,v1)
plt.show()

plt.figure()
plt.plot(frec, np.abs(y1))
plt.axvline(45*2)
plt.xlim([5, 100])
plt.show()
```

Como estamos graficando un valor absoluto se ve una frecuencia en 0 porque los datos no están centrados en 0 y nos aparece una transformada en 0 por el offset del modulo.Y por esto da reflejada la transformada de Fourier. Nos interesa solo la componente que esta en dos veces la frecuencia. 

```python
y2 = np.fft.fft(v2)
y3 = np.fft.fft(v3)
fig, ax = plt.subplots(2, figsize = (20,10))
ax[0].plot(frec, np.abs(y2))
ax[0].set_title('Transformada de Fourier de $v_2$')
ax[0].set_xlim([0, 500])
ax[1].plot(frec, np.abs(y3))
ax[1].set_title('transformada de Foruer de $v_3$')
ax[1].set_xlim([0,500])
```

Aca nuevamente tenemos la transformada distinto de 0 en x = 0 nuevamente por el offset, y salvo esta la mayor importancia en la transformada es en f = 90 osea el doble de la frecuencia y esto es por aplicar el modulo(doble frecuencia). Lo que si en la segunda se repiten las frecuencias
